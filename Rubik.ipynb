**Muestra cuatro imágenes**

```
# Muestra cuatro imágenes
def display4(a, b, c, d, title1 = "Original", title2 = "Edited", title3 = "Edited 3", title4 = "Edited 4"):
  plt.figure(figsize=(14,14))
  plt.subplot(141), plt.imshow(a, cmap='gray'), plt.title(title1)
  plt.xticks([]), plt.yticks([])
  plt.subplot(141), plt.imshow(b, cmap='gray'), plt.title(title2)
  plt.xticks([]), plt.yticks([])
  plt.subplot(143), plt.imshow(c, cmap='gray'), plt.title(title3)
  plt.xticks([]), plt.yticks([])
  plt.subplot(144), plt.imshow(d, cmap='gray'), plt.title(title4)
  plt.xticks([]), plt.yticks([])
  plt.show()
```

# BGR Color-Space

```
# Leer una imagen
# Principle of least astonishment (Surprise)
img_BGR = cv2.imread('ruby.jpg', cv2.IMREAD_COLOR)
#b,g,r = cv2.split(img_BGR)

b = img_BGR[:,:,0]
g = img_BGR[:,:,1]
r = img_BGR[:,:,2]

display4(img_BGR,b,g,r,"BGR","b","g","r")
```
